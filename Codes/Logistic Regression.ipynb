```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("ads.csv")
data
```

```python
#Convert Gender to binary
data['Gender'] = data['Gender'].astype('category').cat.codes
data
```

```python
"""Did not allot x and y before hand because reshaping becomes an issue and when done this way we do not need to
Reshape the array of values"""
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data[['Age']], data.Purchased, test_size=0.2, random_state=21)
x_train, x_test, y_train, y_test
```

```python
#Same as LR and MLR
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Again the same
model.score(x_test,y_test)
predictions = model.predict(x_test)
model.predict([[23]])
```

```python
#Easy just call confusion matrix
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```
